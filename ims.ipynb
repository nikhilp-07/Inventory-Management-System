# The below code is for 
### Addition and Updation = True
### Deletion = False

## 1)adding a new Item/product in to database/Inventory or to update a data of an existing product(only quantity and price).
### i) if product exist we can update the existing data(only quantity and price)
##### - enter quantity for update
##### - enter change in price  for update ( increase = True ,decrease = False)
### ii)else if dosen't exist we can add products in to Inventory.
#### Earlier in database : Total products were = 18; for ex let's add 12 more new  products  in  to database

## 2) deletion of an item/product from the database/Inventory
### i)enter the prod_id to delete the existing product from Inventory
### ii) else If product doesn't exist then it displays a message saying - Entered item prod_id-1100 doesn't exist.


```python
function = input("enter True for Addition and Updation or False for Deletion of product :")
#Addition and Updation = True
#Deletion = False
if function == str(True):
    prod_id = str(input("Enter product id:"))
    if (prod_id in record.keys()):
        print("Product Exist")
        qn = int(input("Enter quantity for update :"))
        record[prod_id]['quantity'] = record[prod_id]['quantity'] + qn 
        pr = int(input("Enter change in price for update  :"))
        i_or_d = input("enter True for increase and False for decrease :")
        #increase = True
        #decrease = False
        if  i_or_d == str(True) :
            record[prod_id]['price'] = record[prod_id]['price'] + pr
        elif i_or_d == str(False)  :
            record[prod_id]['price'] = record[prod_id]['price'] - pr
        else:
            pass
        print("Product data is updated")
    else:
        name = str(input("Enter name:"))
        pr = int(input("Enter price:"))
        qn = int(input("Enter quantity:"))
        caty = str(input("Enter category:"))
        rat = int(input("Enter rating of the product:"))
        record[prod_id] = {'name': name , 'price': pr , 'quantity': qn , 'category':caty , 'customer_rating_out_of_5': rat}
        print("New product successfully added into database/Inventory ")

else:
    prod_id = str(input("Enter product id:"))
    if (prod_id in record.keys()):
        del record[prod_id]
        print("Item is deleted successfully")
    else:
        print(f"Entered item prod_id-{prod_id} doesn't exist")
    
db = json.dumps(record)

fd = open("record.json",'w')
fd.write(db)
fd.close()
```

#  purchase  : user can purchase the product by entering the product_id and quantity
## i) if the (Availability)quantity of required product is 0 in Inventory/database then,  message will be displayed saying - Product is out of stock,Apologize for the Invoncenience and product will be back in stock soon
## ii)else if the quantity of products required by user is more than existing quantity of product in database,then message will be displayed saying - Product is out of stock,Apologize for the Invoncenience and have the quantity of   35 left for prod_ID - 1008 
## iii)generation of bill 
## iv) creating a sales.json file for recording the total products sold and it's basic details like - product_id,quantity,total amount,date-and-time_details,and user phone number.
## v) in the end updating the Inventory/database products.


```python
sale = {}
```

```python
import time
t= time.ctime()

ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))
ui_ph = int(input("enter the phone number :"))

if( record[ui_prod]['quantity'] <= 0):
    print("Product is out of stock,Apologize for the Invoncenience and product will be back in stock soon")
elif( record[ui_prod]['quantity'] < ui_quant ) :
    Availability_of_quantity = record[ui_prod]['quantity']
    print(f"Product is out of stock,Apologize for the Invoncenience and have the quantity of {Availability_of_quantity} left for prod_ID - {ui_prod} ")
else:
    print("******************************************************************")
    print("Product Name   : ", record[ui_prod]['name'])
    print("Price          : ", record[ui_prod]['price'])
    print("------------------------------------------------------------------")
    print("Total Amount   : ", record[ui_prod]['price'] * ui_quant)
    print("------------------------------------------------------------------")
    print("Date of puchase: ",t)
    print("Thank you for shopping with us,have a good day")
    print("******************************************************************")
    
    sale[(len(sale)+1)] = {'product_id': ui_prod , 'quantity': ui_quant , 'amount':record[ui_prod]['price']*ui_quant , 'date-and-time_details':t , 'phone_number': ui_ph}
    xyz=json.dumps(sale)
    
    cd = open("sales.json","w")
    cd.write(xyz)
    cd.close()
    
    
    #updating Inventoty/database
    record[ui_prod]['quantity'] = record[ui_prod]['quantity'] - ui_quant
    

db = json.dumps(record)

fd = open("record.json",'w')
fd.write(db)
fd.close()
```
